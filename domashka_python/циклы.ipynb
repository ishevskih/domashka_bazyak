4,1: Циклы с Заданным числом потворений

```python
for x in range(41):
    n = x**2 + x + 41
    prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            prime = False
            break
    if prime:
        print(f"x = {x}, {n} - простое число")
    else:
        print(f"x = {x}, {n} - не простое число")
```

4.3 Итерационные циклы 


```python
# Функция для вычисления факториала
def factorial(k):
    result = 1
    for i in range(2, k + 1):
        result *= i
    return result

# Ввод точности
eps = float(input("Введите значение ε: "))

# Начальные параметры
n = 1
total = 0.0

# Суммирование членов, модуль которых >= eps
while True:
    # Вычисление a_n = (3^n * n!) / (2n)!
    a_n = (3**n * factorial(n)) / factorial(2 * n)
    
    if abs(a_n) >= eps:
        total += a_n
        n += 1
    else:
        break

# Вывод результата
print(f"Сумма членов ряда (|a_n| >= {eps}) = {total}")

```

4.4 Табулирование функций 

```python
from math import *
from pandas import *

# Ввод параметров
a = float(input("Введите начало отрезка a: "))
b = float(input("Введите конец отрезка b: "))
h = float(input("Введите шаг h: "))

# Генерация значений x
n = int(round((b - a) / h)) + 1
x_values = [a + i * h for i in range(n)]

# Вычисление y = x·cos(1/x) + 2
y_values = []
for x in x_values:
    if abs(x) < 1e-12:
        y = 2.0
    else:
        y = x * cos(1.0 / x) + 2
    y_values.append(round(y, 6))

# Создание DataFrame
df = DataFrame({
    "x": [round(x, 6) for x in x_values],
    "F(x) = x·cos(1/x) + 2": y_values
})

# Вывод таблицы средствами pandas (без псевдографики)
print("\nТабуляция функции F(x) = x·cos(1/x) + 2")
print(df.to_string(index=False))
```
