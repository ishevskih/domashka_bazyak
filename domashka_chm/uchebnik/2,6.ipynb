Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Функция
def f(x):
    return np.log10(x) - np.cos(x)

# График
x = np.linspace(0.1, 3, 500)
plt.plot(x, f(x), 'b-')
plt.axhline(0, color='r', ls='--')
plt.grid(True)
plt.title('lg x - cos x = 0')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

# Метод половинного деления
a, b = 1.3, 1.5  # интервал из графика
for _ in range(30):
    c = (a + b) / 2
    if f(a) * f(c) < 0:
        b = c
    else:
        a = c
    if (b - a) < 1e-5:
        break

print(f"Корень: {c:.4f}")
```

Задание 2

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 3.2*x - 1
a, b = 3.0, 3.5
for _ in range(30):
    c = (a + b) / 2
    if f(a) * f(c) < 0:
        b = c
    else:
        a = c
    if (b - a) < 1e-5:
        break
x = np.linspace(-1, 5, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = x² - 3.2x - 1')
plt.axhline(0, color='r', linestyle='--', label='y = 0')
plt.axvline(0, color='gray', alpha=0.5)
plt.grid(True, alpha=0.3)
plt.plot(c, 0, 'ro', markersize=8, label=f'Корень: x = {c:.4f}')
plt.plot([a, b], [0, 0], 'g|', markersize=12, label='Интервал')
plt.title('Уравнение x² - 3.2x = 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

print(f"Корень: {c:.4f}")
```
