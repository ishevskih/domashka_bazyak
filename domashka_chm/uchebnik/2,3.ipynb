Задание 1

```python
from numpy import *          
from matplotlib.pyplot import *  


def f(x):
    return 0.5*(1 - x) + 0.5

# Создаём 400 точек на отрезке [-0.5, 2] для построения графиков
x = linspace(-0.5, 2, 400)

# Начальное приближение для итерационного процесса
x0 = 1.8
# Количество итераций, которое будем отображать
iterations = 8

# Создаём новое окно для графика
figure()

# Строим прямую y = x (биссектрису первого координатного угла)
plot(x, x, label="y = x")
# Строим график функции y = f(x)
plot(x, f(x), label="y = f(x)")

# Начинаем итерационный процесс с точки x0
xn = x0
for i in range(iterations):
    # Вычисляем значение функции в текущей точке
    yn = f(xn)
    # Рисуем вертикальную линию от (xn, xn) до (xn, yn)
    plot([xn, xn], [xn, yn])
    # Рисуем горизонтальную линию от (xn, yn) до (yn, yn)
    plot([xn, yn], [yn, yn])
    # Переходим к следующей итерации: новое xn становится равным yn
    xn = yn

# Заголовок графика
title("Сходящийся итерационный процесс при -1 < f'(x) < 0")
# Подписи осей
xlabel("x")
ylabel("y")
# Показываем легенду
legend()
# Включаем сетку
grid()

# Отображаем график на экране
show()
```

```python

from numpy import *
from pandas import *

print("Решение уравнения x·sin(x) - 1 = 0")
print("Метод простой итерации, точность 10⁻⁵")

# Функция φ(x) для итерационного процесса: x - 0.5*(x*sin(x) - 1)
def phi(x):
    return x - 0.5 * (x * sin(x) - 1)

# Начальное приближение
x = 1.1

# Список для хранения данных таблицы итераций
data = []

# Выполняем итерации (максимум 20)
for n in range(1, 20):
    x_new = phi(x)               # новое приближение
    diff = abs(x_new - x)         # разница между приближениями

    # Добавляем строку в таблицу
    data.append({
        "n": n,
        "x_n": round(x_new, 8),
        "|x_n - x_n-1|": round(diff, 8)
    })
    
    # Если достигнута требуемая точность, запоминаем корень и выходим
    if diff < 1e-5:
        root = x_new
        break
        
    x = x_new                     # готовимся к следующей итерации

# Создаём DataFrame из собранных данных и выводим таблицу
df = DataFrame(data)
print("\nТаблица итераций:")
print(df.to_string(index=False))

# Вывод результата
print(f"\nКорень: x = {root:.8f}")
print(f"Проверка: {root:.8f}·sin({root:.8f}) - 1 = {root * sin(root) - 1:.2e}")
```
