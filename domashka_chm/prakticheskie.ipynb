```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return np.sin(x)

def f2(x):
    return 0.2 * x
x = np.linspace(-10, 10, 1000)

y1 = f1(x)
y2 = f2(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='sin(x)', linewidth=2)
plt.plot(x, y2, label='0.2x', linewidth=2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графическое отделение корней уравнения sin(x) - 0.2x = 0')
plt.legend()

f = y1 - y2
sign_changes = np.where(np.diff(np.sign(f)))[0]

print("Интервалы, содержащие корни:")
for idx in sign_changes:
    x1, x2 = x[idx], x[idx+1]
    print(f"[{x1:.3f}, {x2:.3f}]")
    plt.axvspan(x1, x2, alpha=0.2, color='yellow')

plt.xlim(-10, 10)
plt.ylim(-2, 2)
plt.show()
```

```python
import math
import pandas as pd

def f(x):
    return math.sin(x) - 0.2 * x

a, b = 2.5, 2.7
eps = 1e-3
data = []
n = 0

while (b - a) > eps:
    c = (a + b) / 2
    fc = f(c)
    data.append([n, a, b, c, fc, b - a])
    if f(a) * fc < 0:
        b = c
    else:
        a = c
    n += 1

df = pd.DataFrame(data, columns=["Итерация", "a", "b", "c", "f(c)", "|b-a|"])
print(df)
print(f"\nКорень: {c:.6f} с точностью {eps}")
```

```python
import math
import pandas as pd

def phi(x):
    return math.pi - math.asin(0.2 * x)

x0 = 2.6
eps = 1e-3
data = []
x = x0
n = 0

while True:
    x_next = phi(x)
    diff = abs(x_next - x)
    data.append([n, x, x_next, diff])
    if diff < eps:
        break
    x = x_next
    n += 1

df = pd.DataFrame(data, columns=["n", "x_n", "x_{n+1}", "|x_{n+1}-x_n|"])
print(df)
print(f"\nКорень: {data[-1][2]:.6f} с точностью {eps}")
```

```python
import math
import pandas as pd

def f(x):
    return math.sin(x) - 0.2 * x

a, b = 2.5, 2.7
eps = 1e-3

x_prev = a
x_curr = b
data = []
n = 0

while True:
    if f(x_curr) - f(x_prev) == 0:
        break
    x_next = x_curr - f(x_curr) * (x_curr - x_prev) / (f(x_curr) - f(x_prev))
    error = abs(x_next - x_curr)
    data.append([n, x_prev, x_curr, x_next, f(x_next), error])
    if error < eps:
        break
    x_prev, x_curr = x_curr, x_next
    n += 1


df = pd.DataFrame(data, columns=["n", "x_prev", "x_curr", "x_next", "f(x_next)", "|x_next-x_curr|"])
print(df)
print(f"\nКорень: {x_next:.6f} с точностью {eps}")
```

```python
import math
import pandas as pd

def f(x):
    return math.sin(x) - 0.2 * x

def df(x):
    return math.cos(x) - 0.2
x0 = 2.5
eps = 1e-3
data = []
x = x0
n = 0

print("Вычисления методом Ньютона для уравнения sin(x) - 0.2x = 0")
print("Начальное приближение: x0 =", x0)
print("Точность: eps =", eps)
print()

while True:
    fx = f(x)
    dfx = df(x)
    x_next = x - fx / dfx
    diff = abs(x_next - x)
    data.append([n, x, fx, dfx, x_next, diff])
    if diff < eps:
        break
    x = x_next
    n += 1
columns = ["n", "x_n", "f(x_n)", "f'(x_n)", "x_{n+1}", "|x_{n+1}-x_n|"]
df_table = pd.DataFrame(data, columns=columns)

print(df_table.to_string(index=False))
print(f"\nКорень уравнения: {x_next:.6f} с точностью {eps}")
```
